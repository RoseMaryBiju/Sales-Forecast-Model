```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.arima.model import ARIMA
from prophet import Prophet
from sklearn.metrics import mean_squared_error, mean_absolute_error
from google.colab import files
```

```python
df = pd.read_csv('train.csv', parse_dates=['Order Date', 'Ship Date'], index_col='Order Date')
df.index = pd.to_datetime(df.index, dayfirst=True) # Tell pandas to interpret the day as coming first
# Data Pre-processing
df.dropna(inplace=True)
df_grouped = df.resample('D').sum()
```

```python
# EDA
plt.figure(figsize=(10, 6))
plt.plot(df_grouped.index, df_grouped['Sales'], label='Daily Sales')
plt.title('Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.show()

decomposition = seasonal_decompose(df_grouped['Sales'], model='additive')
decomposition.plot()
plt.show()
```

```python

# Model Selection and Training
arima_model = ARIMA(df_grouped['Sales'], order=(5, 1, 0))
arima_result = arima_model.fit()
df_grouped['arima_forecast'] = arima_result.predict(start=len(df_grouped), end=len(df_grouped)+30, typ='levels')

prophet_df = df_grouped.reset_index().rename(columns={'Order Date': 'ds', 'Sales': 'y'})
prophet_model = Prophet()
prophet_model.fit(prophet_df)
future = prophet_model.make_future_dataframe(periods=30)
forecast = prophet_model.predict(future)
df_grouped['prophet_forecast'] = forecast.set_index('ds')['yhat']
```

```python
# Model Selection and Training
arima_model = ARIMA(df_grouped['Sales'], order=(5, 1, 0))
arima_result = arima_model.fit()

# Generating predictions for the entire date range of df_grouped
# Using the index of df_grouped to ensure alignment
predictions = arima_result.predict(start=df_grouped.index.min(), end=df_grouped.index.max(), typ='levels')

# Assigning the predictions to the DataFrame, aligning on the index
df_grouped['arima_forecast'] = predictions

prophet_df = df_grouped.reset_index().rename(columns={'Order Date': 'ds', 'Sales': 'y'})
prophet_model = Prophet()
prophet_model.fit(prophet_df)
future = prophet_model.make_future_dataframe(periods=30)
forecast = prophet_model.predict(future)
df_grouped['prophet_forecast'] = forecast.set_index('ds')['yhat']
```

```python
# Plotting
plt.figure(figsize=(10, 6))
plt.plot(df_grouped.index, df_grouped['Sales'], label='Actual Sales')
plt.plot(df_grouped.index, df_grouped['arima_forecast'], label='ARIMA Forecast')
plt.plot(df_grouped.index, df_grouped['prophet_forecast'], label='Prophet Forecast')
plt.title('Sales Forecast')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

```python
import numpy as np
# Model Evaluation
rmse_arima = np.sqrt(mean_squared_error(df_grouped['Sales'], df_grouped['arima_forecast'].dropna()))
mae_arima = mean_absolute_error(df_grouped['Sales'], df_grouped['arima_forecast'].dropna())

rmse_prophet = np.sqrt(mean_squared_error(df_grouped['Sales'], df_grouped['prophet_forecast'].dropna()))
mae_prophet = mean_absolute_error(df_grouped['Sales'], df_grouped['prophet_forecast'].dropna())

# results
print(f'ARIMA RMSE: {rmse_arima}, MAE: {mae_arima}')

print(f'Prophet RMSE: {rmse_prophet}, MAE: {mae_prophet}')
```

```python
from sklearn.model_selection import train_test_split

# Assuming df_grouped is the DataFrame with the resampled sales data
# Splitting the data into training and testing sets
train_size = int(len(df_grouped) * 0.8)
train, test = df_grouped.iloc[:train_size], df_grouped.iloc[train_size:]
```

```python
from statsmodels.tsa.stattools import adfuller

# Performing Augmented Dickey-Fuller test
result = adfuller(train['Sales'])
print('ADF Statistic:', result[0])
print('p-value:', result[1])
for key, value in result[4].items():
    print('Critical Value (%s): %.3f' % (key, value))
```

```python
import itertools
import warnings
warnings.filterwarnings("ignore")

# Defining the p, d, q parameters to take any value between 0 and 2
p = d = q = range(0, 3)
pdq = list(itertools.product(p, d, q))

best_aic = float("inf")
best_params = None

for param in pdq:
    try:
        arima_model = ARIMA(train['Sales'], order=param)
        results = arima_model.fit()
        if results.aic < best_aic:
            best_aic = results.aic
            best_params = param
    except:
        continue

print('Best ARIMA parameters:', best_params)
```

```python
# Fitting the ARIMA model with the best parameters
arima_model = ARIMA(train['Sales'], order=best_params)
arima_result = arima_model.fit()

# Forecast
arima_forecast = arima_result.forecast(steps=len(test))
```

```python
# Preparing the data for Prophet
prophet_train = train.reset_index().rename(columns={'Order Date': 'ds', 'Sales': 'y'})
prophet_test = test.reset_index().rename(columns={'Order Date': 'ds', 'Sales': 'y'})

# Fitting the Prophet model
prophet_model = Prophet()
prophet_model.fit(prophet_train)

#  future dataframe for forecast
future = prophet_model.make_future_dataframe(periods=len(test))
forecast = prophet_model.predict(future)

# Extracting the forecasted values
prophet_forecast = forecast.set_index('ds').loc[test.index]['yhat']
```

```python
#  RMSE, MAE, MAPE for ARIMA
rmse_arima = np.sqrt(mean_squared_error(test['Sales'], arima_forecast))
mae_arima = mean_absolute_error(test['Sales'], arima_forecast)
mape_arima = np.mean(np.abs(arima_forecast - test['Sales']) / np.abs(test['Sales'])) * 100

print(f'ARIMA RMSE: {rmse_arima}, MAE: {mae_arima}, MAPE: {mape_arima}')

#  RMSE, MAE, MAPE for Prophet
rmse_prophet = np.sqrt(mean_squared_error(test['Sales'], prophet_forecast))
mae_prophet = mean_absolute_error(test['Sales'], prophet_forecast)
mape_prophet = np.mean(np.abs(prophet_forecast - test['Sales']) / np.abs(test['Sales'])) * 100

print(f'Prophet RMSE: {rmse_prophet}, MAE: {mae_prophet}, MAPE: {mape_prophet}')
```

```python
# Residuals for ARIMA
residuals_arima = test['Sales'] - arima_forecast
plt.figure(figsize=(10, 6))
plt.plot(residuals_arima)
plt.title('ARIMA Residuals')
plt.show()

# Residuals for Prophet
residuals_prophet = test['Sales'] - prophet_forecast
plt.figure(figsize=(10, 6))
plt.plot(residuals_prophet)
plt.title('Prophet Residuals')
plt.show()
```

```python
# Plotting Actual vs Forecasted Sales
plt.figure(figsize=(10, 6))
plt.plot(test.index, test['Sales'], label='Actual Sales')
plt.plot(test.index, arima_forecast, label='ARIMA Forecast')
plt.plot(test.index, prophet_forecast, label='Prophet Forecast')
plt.title('Sales Forecast Comparison')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.show()
```
